```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
a=[]
for i in range(1,101):
    url = "https://www.startupwala.com/list-of-registered-companies-in-india-P"+str(i)
    response= requests.get(url)
    html_text=response.text
    soup= BeautifulSoup(html_text,'html.parser')
    tag_a= soup.find_all('a')
    a.append(tag_a)
```

```python
a
```

```python
link=[]
for i in a:
    for j in i:
        link.append(j.get('href'))
l=list(set(link))
```

```python
l
```

```python
pg_link=[]
for i in l:
    if i[0]=="c":
        pg_link.append(i)
pg_link
```

```python
p_link=['https://www.startupwala.com/'+i for i in pg_link]
```

```python
p_link
```

```python
company_name = []
CIN= []
ROC_Name= []
Date_of_Incorporation= []
Registered_Address= []
Authorised_Capital= []
Paid_Up_Capital= []
Date_of_Last_AGM= []
Date_of_Latest_Balance_sheet= []
Company_Status= []
for i in pg_link:
    url="https://www.startupwala.com/"+str(i)
    response=requests.get(url)
    html_text=response.text
    soup=BeautifulSoup(html_text,"html.parser")
    company_name.append(soup.find_all("td")[1].text.strip())
    CIN.append(soup.find_all("td")[3].text.strip())
    ROC_Name.append(soup.find_all("td")[5].text.strip())
    Date_of_Incorporation.append(soup.find_all("td")[7].text.strip())
    Registered_Address.append(soup.find_all("td")[9].text.strip())
    Authorised_Capital.append(soup.find_all("td")[11].text.strip())
    Paid_Up_Capital.append(soup.find_all("td")[13].text.strip())
    Date_of_Last_AGM.append(soup.find_all("td")[15].text.strip())
    Date_of_Latest_Balance_sheet.append(soup.find_all("td")[17].text.strip())
    Company_Status.append(soup.find_all("td")[19].text.strip())
```

```python
p = pd.DataFrame({"company Name":company_name,'CIN':CIN,"ROC Name":ROC_Name,"Date of Incorporation":Date_of_Incorporation,
                         "Registered Address":Registered_Address,
                         "Authorised Capital":Authorised_Capital,"Paid Up Capital":Paid_Up_Capital,
                         "Date of Last AGM":Date_of_Last_AGM,"Date of Latest Balance sheet":Date_of_Latest_Balance_sheet,
                        "Company Status":Company_Status})
```
